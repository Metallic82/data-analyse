```python
#Controlling axes of Grid
h = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre' )
kws = dict(s=50,linewidth=0.5, edgecolor='black')
h = h.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
h.set(xlim=(0,100) , ylim=(0,100)) #Axes Length
for ax in h.axes.flat:
    ax.plot( (0,100) , (0,100) , c='grey' , ls = '--')
h.add_legend()
```

Styling Dashboards

```python
sns.set_style('dark' , {'axes.facecolor' :'black'})
f , axes = plt.subplots( 2 , 2 , figsize = (16,16) )

#Plot [0,0]
k3 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating, shade = True , shade_lowest= False ,\
                 cmap = 'inferno' ,ax=axes[0,0])
k3b = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating, \
                 cmap = 'Blues' ,ax=axes[0,0])
#Plot [0,1]
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating, shade = True , shade_lowest = True , \
                 cmap = 'inferno',ax=axes[0,1])
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating, \
                 cmap = 'Blues',ax=axes[0,1])
#Plot [1,0]
g = sns.violinplot(data = movies, x = 'Year', y = 'BudgetMillions', \
                   palette = 'YlOrRd' , ax=axes[1,0])
#Plot [1,1]
k1 = sns.kdeplot(movies.CriticRating , movies.AudienceRating, shade=True, shade_lowest=False, gridsize=100, cmap='Blues_r', ax=axes[1,1])
k1b = sns.kdeplot(movies.CriticRating , movies.AudienceRating, gridsize=100, cmap='gist_gray_r', ax=axes[1,1])
k3.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
#Thematic Edits
sns.set_style('whitegrid') #Seaborn works even if plot is of mpl
plt.figure(figsize=(12,6))
fig , ax = plt.subplots()
fig.set_size_inches(11.7,8.27)
plt.hist(gen_list,stacked=True,label=my_labels,bins=30)
plt.title("Movie Budget Distribution",fontsize=35 , color = 'DarkBlue' , fontname='Console')
plt.ylabel("Number of Movies" , fontsize= 15,color='DarkRed')
plt.xlabel("$ Millions" , fontsize= 15,color='DarkGreen')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(frameon = True , fancybox = True ,prop={'size':20} , shadow = True , framealpha = 1) #Legend Size
plt.show()
```
